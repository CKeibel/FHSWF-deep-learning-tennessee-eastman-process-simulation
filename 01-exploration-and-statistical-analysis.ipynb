# Statistical detection

```python
!dir data
```

```python
test_ff =  pyreadr.read_r("data/TEP_FaultFree_Testing.RData")
test_f = pyreadr.read_r("data/TEP_Faulty_Testing.RData")
```

```python
test_f.keys(), test_ff.keys()
```

```python
test_ff = test_ff["fault_free_testing"]
test_f = test_f["faulty_testing"]
```

```python
from tqdm import tqdm
import warnings

warnings.filterwarnings('ignore', category=UserWarning)

data = []

with tqdm(total=len(test_f["simulationRun"].unique())) as pbar:
    for run_num in test_f["simulationRun"].unique():
        _df = test_f[test_f["simulationRun"] == run_num]
        x = _df.iloc[:, 3:]
        y = _df["faultNumber"].to_list()[0]
        data.append({"x":x, "y":y})
        pbar.update(1)

with tqdm(total=len(test_ff["simulationRun"].unique())) as pbar:
    for run_num in test_ff["simulationRun"].unique():
        _df = test_ff[test_ff["simulationRun"] == run_num]
        x = _df.iloc[:, 3:]
        y = _df["faultNumber"].to_list()[0]
        data.append({"x":x, "y":y})
        pbar.update(1)
```

```python
# mean, stddev, upper limit, lower limit
mean = test_ff.mean(axis=0)
std = test_ff.std(axis=0)

upper_limit = mean+4*std
lower_limit = mean-4*std

def is_anomaly(sample, key):
    result = ((sample < lower_limit[key]).any()) or ((sample > upper_limit[key]).any())
    return result
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sn

features = list(test_f.columns[3:])

false_positives = {}
false_negatives = {}
correct = []
anomalies_count = 0
no_anomalies_count = 0
y_pred = []
y_true = []

with tqdm(total=len(data)) as pbar:
    for idx, sample in enumerate(data):
        x = sample["x"]
        y = sample["y"]
        
        anomalies = []
        for f in features:
            result = is_anomaly(x[f], f)
            if result:
                anomalies.append(f)
        y = int(y)
        if len(anomalies) > 0 and y == 0:
            false_positives[idx] = anomalies
        elif len(anomalies) == 0 and y > 0:
            false_negatives[idx] = anomalies
        else:
            correct.append(sample)

        if len(anomalies) > 0:
            anomalies_count += 1
        else:
            no_anomalies_count += 1

        y_pred.append(1 if len(anomalies) > 0 else 0)
        y_true.append(1 if y > 0 else 0)
    
        pbar.update(1)


cm = confusion_matrix(y_true, y_pred)
df_cm = pd.DataFrame(cm, index=["Normal", "Anomaly"], columns=["Normal", "Anomaly"])
sn.heatmap(df_cm, annot=True, fmt=".0f")
plt.show()

print(f"Correct: {len(correct)}/{len(data)} - {(len(correct)/len(data))*100}% with {anomalies_count} anomalies and {no_anomalies_count} no anomalies.")
print(f"False positives: {len(false_positives)}/{len(data)} - {(len(false_positives)/len(data))*100}%")
print(f"False negatives: {len(false_negatives)}/{len(data)} - {(len(false_negatives)/len(data))*100}%")
```
